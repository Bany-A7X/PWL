```python
from selenium import webdriver
from selenium.webdriver import Chrome
from selenium.webdriver.common.keys import Keys
import time

browser = webdriver.Chrome(executable_path='C:/chromedriver.exe')

browser.get('https://web-otomasi.herokuapp.com/')
print(browser.title)

#case 1: field nama kosong
browser.find_element_by_name("firstName").send_keys("")
time.sleep(2)
browser.find_element_by_name("userEmail").send_keys("BanyA7X@gmail.com")
time.sleep(2)
browser.find_element_by_name("userPassword").send_keys("ABCde22")
time.sleep(2)
browser.find_element_by_name("confirmPassword").send_keys("ABCde22")
time.sleep(2)
button = browser.find_element_by_tag_name('button')
button.click()
alert_obj = browser.switch_to.alert
time.sleep(2)
alert_obj.accept () 
time.sleep(2)
browser.refresh()

#case 2: field email tidak sesuai aturan
browser.find_element_by_name("firstName").send_keys("Bany Majesty")
time.sleep(2)
browser.find_element_by_name("userEmail").send_keys("bany")
time.sleep(2)
browser.find_element_by_name("userPassword").send_keys("ABCde22")
time.sleep(2)
browser.find_element_by_name("confirmPassword").send_keys("ABCde22")
time.sleep(2)
#memperbaiki penulisan email dengan menambahkan kata "@gmail.com"
browser.find_element_by_name("userEmail").send_keys("@gmail.com")
time.sleep(2)
button = browser.find_element_by_tag_name('button')
button.click()
alert_obj = browser.switch_to.alert
time.sleep(2)
alert_obj.accept () 
time.sleep(2)
browser.refresh()

#case 3: field password tidak sesuai aturan
browser.find_element_by_name("firstName").send_keys("Bany Majesty")
time.sleep(2)
browser.find_element_by_name("userEmail").send_keys("BanyA7X@gmail.com")
time.sleep(2)
browser.find_element_by_name("userPassword").send_keys("abcde22")
time.sleep(2)
#memblok semua teks yang telah diketik dalam field
browser.find_element_by_name("userPassword").send_keys(Keys.CONTROL + "a")
time.sleep(2)
#memencet tombol delete pada keyboard
browser.find_element_by_name("userPassword").send_keys(Keys.DELETE)
time.sleep(2)
#menulis kembali field dengan isi yang benar
browser.find_element_by_name("userPassword").send_keys("ABCde22")
time.sleep(2)
browser.find_element_by_name("confirmPassword").send_keys("ABCde22")
time.sleep(2)
button = browser.find_element_by_tag_name('button')
button.click()
alert_obj = browser.switch_to.alert
time.sleep(2)
alert_obj.accept () 
time.sleep(2)
browser.refresh()

#case 4: field re-password tidak cocok
browser.find_element_by_name("firstName").send_keys("Bany Majesty")
time.sleep(2)
browser.find_element_by_name("userEmail").send_keys("BanyA7X@gmail.com")
time.sleep(2)
browser.find_element_by_name("userPassword").send_keys("ABCde22")
time.sleep(2)
browser.find_element_by_name("confirmPassword").send_keys("abcde22")
time.sleep(2)
#memblok semua teks yang telah diketik dalam field
browser.find_element_by_name("confirmPassword").send_keys(Keys.CONTROL + "a")
time.sleep(2)
#memencet tombol delete pada keyboard
browser.find_element_by_name("confirmPassword").send_keys(Keys.DELETE)
time.sleep(2)
#menulis kembali field dengan isi yang benar
browser.find_element_by_name("confirmPassword").send_keys("ABCde22")
time.sleep(2)
button = browser.find_element_by_tag_name('button')
button.click()
alert_obj = browser.switch_to.alert
time.sleep(2)
alert_obj.accept () 
time.sleep(2)
browser.refresh()

#case 5: semua field lengkap
browser.find_element_by_name("firstName").send_keys("Bany Majesty")
time.sleep(2)
browser.find_element_by_name("userEmail").send_keys("BanyA7X@gmail.com")
time.sleep(2)
browser.find_element_by_name("userPassword").send_keys("ABCde22")
time.sleep(2)
browser.find_element_by_name("confirmPassword").send_keys("ABCde22")
time.sleep(2)
button = browser.find_element_by_tag_name('button')
button.click()
alert_obj = browser.switch_to.alert
time.sleep(2)
alert_obj.accept () 
time.sleep(2)
browser.refresh()
```
